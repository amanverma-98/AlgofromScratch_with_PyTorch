```python
import pandas as pd
```

```python
# Data Preparation
df = pd.read_csv('HeartDiseaseTrain-Test.csv')
df.head()
```

```python
#Data Cleaning
#One-Hot Encoding
df = pd.get_dummies(df , columns = ['sex' , 'chest_pain_type' , 'fasting_blood_sugar','rest_ecg' , 'exercise_induced_angina' , 'slope' , 'vessels_colored_by_flourosopy', 'thalassemia'] , drop_first=True)
df.head()
```

```python
df = df.replace({True: 1, False: 0})
df.head()
```

```python
cols_to_drop = [
    "fasting_blood_sugar_Lower than 120 mg/ml",
    "rest_ecg_Normal",
    "rest_ecg_ST-T wave abnormality",
    "slope_Flat",
    "slope_Upsloping",
    "vessels_colored_by_flourosopy_Zero",
    "vessels_colored_by_flourosopy_One",
    "vessels_colored_by_flourosopy_Two",
    "vessels_colored_by_flourosopy_Three",
    "thalassemia_No",
    "thalassemia_Normal",
    "thalassemia_Reversable Defect"
]
```

```python
X = df.drop(cols_to_drop + ["target"], axis=1).values       # values is used to convert in numpy array
y = df["target"].values.reshape(-1, 1).astype("float32")    # reshaping is done to match with the y_pred shape as sigmoid give y_pred in shape (n_samples , 1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#Standardization
scaler = StandardScaler()
X = scaler.fit_transform(X)

y_svm = (y * 2) - 1   # svm k output col ko [-1 , 1] m badal rhe h
```

```python
print(df.columns)
```

```python
#Train test split
X_train , X_test , y_train , y_test , y_svm_train , y_svm_test = train_test_split(X , y , y_svm , test_size = 0.2 , random_state = 42)
```

```python
import torch
from torch.utils.data import Dataset , DataLoader
```

```python
class CustomDataset(Dataset):
  def __init__(self, X , y , y_svm):
    self.X = torch.tensor(X , dtype = torch.float32)
    self.y = torch.tensor(y , dtype = torch.float32)
    self.y_svm = torch.tensor(y_svm , dtype = torch.float32)

  def __len__(self):
    return len(self.X)

  def __getitem__(self, index):
    return self.X[index] , self.y[index] , self.y_svm[index]
```

```python
#CustomDataset
train_dataset = CustomDataset(X_train , y_train , y_svm_train)
test_dataset = CustomDataset(X_test , y_test , y_svm_test)

#DataLoader
train_loader = DataLoader(train_dataset , batch_size = 16 , shuffle = True)
test_loader = DataLoader(test_dataset , batch_size = 16 , shuffle = False)   # stabiliy lane k liye hm shuffle kr rhe h
```

```python
input_dim = X.shape[1]
```

**Logistic Regression**

```python
w = torch.randn(X_train.shape[1], 1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

def sigmoid(z):
    return 1 / (1 + torch.exp(-z))

def logistic_forward(X):
    return sigmoid(X @ w + b)    # @ for matrix multiplication

def logistic_loss(pred, y):
    eps = 1e-9
    return -(y*torch.log(pred+eps) + (1-y)*torch.log(1-pred+eps)).mean()
```

```python
lr = 0.05
epochs = 10

print("\nTraining Logistic Regression")
for epoch in range(epochs):
    total_loss = 0
    for Xb, yb , ysvm in train_loader:

        pred = logistic_forward(Xb)
        loss = logistic_loss(pred, yb)

        loss.backward()
        with torch.no_grad():
            w -= lr * w.grad
            b -= lr * b.grad

        w.grad.zero_()
        b.grad.zero_()     # to remove accumulation of gradients
        total_loss += loss.item()

    if epoch % 2 == 0:
        print(f"Epoch {epoch}  Loss = {total_loss/len(train_loader):.4f}")

# Eval
with torch.no_grad():
    correct, total = 0, 0
    for Xb, yb, ysvm in test_loader:
        preds = (logistic_forward(Xb) >= 0.5).float()   # converting y_pred into 0 and 1
        correct += (preds == yb).sum().item()
        total += len(yb)

print("Logistic Regression Test Accuracy:", correct/total)


with torch.no_grad():
    correct, total = 0, 0
    for Xb, yb, ysvm in train_loader:
        preds = (logistic_forward(Xb) >= 0.5).float()   # converting y_pred into 0 and 1
        correct += (preds == yb).sum().item()
        total += len(yb)

print("Logistic Regression Train Accuracy:", correct/total)
```

**Support Vector Machine**

```python
w_svm = torch.randn(X_train.shape[1], 1, requires_grad=True)
b_svm = torch.zeros(1, requires_grad=True)

def svm_forward(X):
    return X @ w_svm + b_svm

def svm_loss(pred, y, C=1.0):
    hinge = torch.clamp(1 - y*pred, min=0)   # making all values in range
    return 0.5*torch.sum(w_svm**2) + C * hinge.mean()
```

```python
lr = 0.01
epochs = 5

print("\nTraining SVM...")
for epoch in range(epochs):
    total_loss = 0
    for Xb, yb, ysvm in train_loader:

        pred = svm_forward(Xb)
        loss = svm_loss(pred, ysvm)

        loss.backward()
        with torch.no_grad():
            w_svm -= lr * w_svm.grad
            b_svm -= lr * b_svm.grad

        w_svm.grad.zero_()
        b_svm.grad.zero_()
        total_loss += loss.item()

    if epoch % 1 == 0:
        print(f"Epoch {epoch}  Loss = {total_loss/len(train_loader):.4f}")

# Eval
with torch.no_grad():
    correct, total = 0, 0
    for Xb, yb, ysvm in test_loader:
        preds = torch.where(svm_forward(Xb) > 0, 1.0, -1.0)  # making prediction entries 1 or -1
        correct += (preds == ysvm).sum().item()
        total += len(ysvm)

print("SVM Test Accuracy:", correct/total)


with torch.no_grad():
    correct, total = 0, 0
    for Xb, yb, ysvm in train_loader:
        preds = torch.where(svm_forward(Xb) > 0, 1.0, -1.0)  # making prediction entries 1 or -1
        correct += (preds == ysvm).sum().item()
        total += len(ysvm)

print("SVM Train Accuracy:", correct/total)
```
